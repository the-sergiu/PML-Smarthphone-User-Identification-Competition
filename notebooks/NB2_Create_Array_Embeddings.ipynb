```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import pickle
```

```python
train_files = os.listdir('train/train/')
train_files.sort()
train_files
```

# Each axis will have 3 features: avg, std, var -> 9 features per file

## Let's create X_train

```python
colnames = ['x', 'y', 'z']
N = 9000
features = 9
X_train = np.zeros((N, features))
labels = []

for i, file in enumerate(train_files[:N]):
    
    full_path = f'train/train/{file}'
    temp_df = pd.read_csv(full_path, names=colnames)
#     index = map_file_to_label(file)
#     labels.append(train_labels_map[index])
    
    x = temp_df['x'].to_numpy()
    y = temp_df['y'].to_numpy()
    z = temp_df['z'].to_numpy()
    
    mean_x, std_x, var_x = np.mean(x), np.std(x), np.var(x)
    mean_y, std_y, var_y = np.mean(y), np.std(y), np.var(y)
    mean_z, std_z, var_z = np.mean(z), np.std(z), np.var(z)
    row = [mean_x, std_x, var_x, mean_y, std_y, var_y, mean_z, std_z, var_z]
    
    X_train[i] = np.array(row)
```

```python
X_train
```

```python
# Save X_train
with open('X_train_9.npy', 'wb') as f:
    np.save(f, X_train)
```
